```python
import matplotlib.pyplot as plt
import csv

```

```python
x = []
y = []
 
with open("example.txt", "r") as csvfile:
    plots = csv.reader(csvfile ,delimiter= ",")
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
plt.plot(x,y, label = "Loaded from file")

plt.xlabel("x")
plt.ylabel("y")
plt.title("here is a graph")
plt.show()
```
