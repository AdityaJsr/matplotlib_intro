```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [6,4,2,6,2]

x2 = [1,2,3,4,5]
y2 = [15,12,18,22,14]

plt.plot(x, y , label = "First Line")
plt.plot(x2, y2 , label = "Second line")
plt.xlabel("i am x axis")
plt.ylabel("i am y axis")
plt.title("I am the Graph")
plt.legend()
plt.show()
```
